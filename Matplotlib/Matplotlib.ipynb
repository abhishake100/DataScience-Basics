## Importing Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1,2,3,4,5,6,7]
temp = [30,32,45,30,47,54,45]
plt.plot(days, temp)                                   # A simple 2d chart
```

## Labels, Legend and Grid

```python
plt.xlabel('Weeks')
plt.ylabel('Temperature (Degrees)')
plt.title('Weather')
plt.plot(days,temp)
```

```python
temp_usa = [20,19,32,22,20,26,25]

plt.xlabel('Weeks')
plt.ylabel('Temperature (Degrees)')
plt.title('Weather')

plt.plot(days, temp, label='India')
plt.plot(days, temp_usa, label='USA')

plt.legend()

plt.grid()                             # adds grid to the graph
```

## Customizing the Charts

https://matplotlib.org/3.2.2/api/_as_gen/matplotlib.pyplot.plot.html

```python
plt.plot(days, temp, label='India', color='red', linewidth=4, linestyle='dashed')
plt.plot(days, temp_usa, label='USA', color='black', marker='.', markersize=20)

plt.legend()

plt.grid()
```

```python
plt.plot(days, temp, label='India', color='red', linewidth=4, linestyle='dashed')
plt.plot(days, temp_usa, label='USA', color='black', marker='.', markersize=20)

plt.legend(loc='best', fontsize='large')           #loc is location

plt.grid()
```

```python
temp_china = [26,28,30,27,25,22,38]

# You can use hex values for colors and single letter as well.

plt.plot(days, temp, label='India', color='red', linewidth=2, linestyle='dashed')
plt.plot(days, temp_usa, label='USA', color='b', marker='*', markersize=8, linestyle='-.', markerfacecolor='red')
plt.plot(days, temp_china, label='China', color='#7a2b3d')

plt.legend(loc='upper right', fontsize='large')

plt.grid()
```

```python
plt.plot(days, temp, label='India', color='red', linewidth=2, linestyle='dashed')
plt.plot(days, temp_usa, label='USA', color='b', marker='*', markersize=8, linestyle='-.', markerfacecolor='red')
plt.plot(days, temp_china, label='China', color='#7a2b3d')

plt.legend(loc='best')

plt.tight_layout()    # adjusts the padding and overall graph layout
```

## Bar Charts

```python
plt.xlabel('Weeks')
plt.ylabel('Temperature (Degrees)')
plt.title('Weather')

plt.bar(days, temp, label='India')

plt.legend(loc='best') 
```

### Multiple Bar Charts

```python
import numpy as np

x_index = np.arange(len(days))
width = 0.25

plt.xlabel('Weeks')
plt.ylabel('Temperature (Degrees)')
plt.title('Weather')

plt.style.use('fivethirtyeight')

plt.bar(x_index-width, temp, label='India', width=width)
plt.bar(x_index, temp_usa, label='USA', width=width)
plt.bar(x_index+width, temp_china, label='China', width=width)

plt.xticks(x_index, days)
plt.legend(loc='best')
plt.tight_layout()
```

### Horizontal Bar Chart

```python
plt.xlabel('Weeks')
plt.ylabel('Temperature (Degrees)')
plt.title('Weather')
height = 0.50

plt.style.use('fivethirtyeight')

plt.barh(x_index, temp_usa, label='USA', height=height)

plt.legend()
plt.yticks(x_index, days)
plt.tight_layout()
```

## Line Graph and and Reading/Analyzing data from CSV file

```python
import pandas as pd

gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(15,5))       # overall graph size (length, height)
plt.title('Gas Prices')
plt.xlabel('Year')
plt.ylabel('Price (Dollars)')

plt.plot(gas.Year, gas.Canada, label='Canada')
plt.plot(gas.Year, gas.USA, label='USA')
plt.plot(gas.Year, gas['South Korea'], label='South Korea')

plt.xticks(gas.Year)

plt.legend()
plt.grid()
```

```python
# printing the whole data

plt.figure(figsize=(15,6))

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country])

plt.xticks(gas.Year)
plt.grid()
```

```python
ctr = ['Australia','Canada','France','Germany','Italy','Japan','Mexico','South Korea','UK','USA']

plt.figure(figsize=(15,6))

for country in gas:
    if country in ctr:
        plt.plot(gas.Year, gas[country])

plt.xticks(gas.Year)
plt.grid()
```

## Histogram

```python
fifa = pd.read_csv('fifa19.csv')
fifa.head()
```

```python
bins = [30,40,50,60,70,80,90,100]

plt.xlabel('Overall Rating')
plt.ylabel('No. of Players')

plt.hist(fifa.Overall, bins, color='green')
plt.grid()
plt.show()

# Here there are players more than 90 ratings but the numbers of players is low, so can't see it in the graph.
```

```python
bins = [85,90,95,100]

plt.xlabel('Overall Rating')
plt.ylabel('No. of Players')

plt.hist(fifa.Overall, bins, color='#abcdef')
plt.grid()

# Here you can see that there are players above 90 overall ratings.
```

### Multiple Histogram

```python
bins = [20,30,40,50,60,70,80,90,100]

plt.figure(figsize=(15,6))
plt.hist([fifa.Overall, fifa.Potential], bins, label=['Overall','Potential'], color=['lightblue','pink'],
         orientation='horizontal')

plt.legend()
plt.show()
```

## Pie Chart

```python
exp = [1400,600,300,410,250]
exp_items = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]

plt.pie(exp, labels=exp_items)
plt.show()
```

```python
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
labels = ['Right Foot','Left Foot']
plt.pie([right, left], labels=['Right foot', 'Left Foot'], radius=1.5, colors=['red','lightblue'], autopct='%.2f%%')
plt.show()

# autocpt will show us the percentage, the number specified is how many numbers you want after the decimal point.
```

## Box Plot
The middle line in the box is the median. The top line is max value and the bottom line is the min value.

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
realmadrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
everton = fifa.loc[fifa.Club == 'Everton']['Overall']

plt.figure(figsize=(10,5))
plt.title('Football Team Comparison')
plt.ylabel('Overall Rating')

plt.boxplot([barcelona, everton, realmadrid], labels=['FC Barcelona','Everton','Real Madrid'], patch_artist=True)
plt.grid()

plt.show()
```
